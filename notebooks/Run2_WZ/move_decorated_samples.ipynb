```python
### This notebook moves GM signal samples and bkg samples to a new folder
from pathlib import Path
import numpy as np
from shutil import copyfile
import logging

from utils.utils import check_inputpath, check_outputpath, logging_setup
```

```python
input_path = "/home/yolo/Projects/ssWWWZjj_sample/Run2/WZ"
input_path = Path(input_path)
check_inputpath(input_path)

output_path = "../Run2_WZ_samples"
output_path = Path(output_path)
check_outputpath(output_path)
```

```python
ntuples_path = input_path / "ntuples" / "main"
```

```python
GM_sig_folder_patterns = [
    "*aMCPy8EG_GM_H5_WZ_lvll*",
    "*MGaMcAtNloPy8EG_A14NNPDF23LO_vbfGM_sH05*",
]
GM_sig_ntuple_pattern = "resonance*GM*ntuples.root"

GM_sig_folders = []
for GM_sig_pattern in GM_sig_folder_patterns:
    GM_sig_folders.extend(ntuples_path.glob(GM_sig_pattern))

mass_points = np.hstack((np.arange(200, 551, 25, dtype=int), np.arange(600, 1001, 100, dtype=int)))
assert len(GM_sig_folders) == len(mass_points)

GM_sig_path = output_path / "decorated" / "GM_sig"
check_outputpath(GM_sig_path)

for GM_sig_folder in GM_sig_folders:
    GM_sig_folder = GM_sig_folder / "decorated"
    GM_sig_file = sorted(GM_sig_folder.glob(GM_sig_ntuple_pattern))

    if len(GM_sig_file) == 0:
        logging.error(f"No file found in {GM_sig_folder}")
    if len(GM_sig_file) > 1:
        logging.info(f"More than one file found in {GM_sig_folder}")

    if len(GM_sig_file) != 1:
        raise ValueError(f"More than one file found in {GM_sig_folder}")

    GM_sig_file = GM_sig_file[0]

    copyfile(GM_sig_file.as_posix(), GM_sig_path / GM_sig_file.name)
```

```python
bkg_folder_patterns = [
    "*364253_Sherpa_222_NNPDF30NNLO_lllv*",
    "*3647*_MGPy8EG_NNPDF30NLO_A14NNPDF23LO_lvlljjEW6_*",
]
bkg_ntuple_pattern = "resonance*364*ntuples.root"
bkg_folders = []
for bkg_folder_pattern in bkg_folder_patterns:
    bkg_folders.extend(ntuples_path.glob(bkg_folder_pattern))


bkg_path = output_path / "decorated" / "bkg"
check_outputpath(bkg_path)

for bkg_folder in bkg_folders:
    bkg_folder = bkg_folder / "decorated"
    bkg_file = sorted(bkg_folder.glob(bkg_ntuple_pattern))

    if len(bkg_file) == 0:
        logging.error(f"No file found in {GM_sig_folder}")
    if len(bkg_file) > 1:
        logging.info(f"More than one file found in {GM_sig_folder}")

    if len(bkg_file) != 1:
        raise ValueError(f"More than one file found in {GM_sig_folder}")

    bkg_file = bkg_file[0]

    copyfile(bkg_file.as_posix(), bkg_path / bkg_file.name)
```
